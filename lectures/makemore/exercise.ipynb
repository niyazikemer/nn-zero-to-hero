```python
import json
with open('isimler.json', 'r') as f:
    data = json.load(f)
```

```python
names = [person['name'].lower() for person in data]
names,len(names)
```

### Create Bigrams

```python
letter_list = set()
old_chars = ['˜','š','¦',' ']
new_chars = ['i','ü','ğ','']
for name in names: 
    for old_char, new_char in zip(old_chars, new_chars):
        name = name.replace(old_char, new_char)
       
    nm_token = ['.'] + list(name) + ['.']
    for ch1,ch2 in zip(nm_token,nm_token[1:]):
        # if '' in ch1 or '' in ch2 :
        #     print(name)
        letter_list.add(ch1)
        letter_list.add(ch2)
        

letter_list
```

### Create int to string and string to int dictionaries

```python

itos  = { no+1:letter for no, letter in enumerate(letter_list)}
itos
```
